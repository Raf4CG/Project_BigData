```python
import networkx as nx
import numpy as np
```

```python
distancias = [[0, 18, 32, 0, 0, 0],
              [18, 0, 28, 12, 0, 0],
              [32, 28, 0, 17, 4, 17],
              [0, 12, 17, 0, 0, 32],
              [0, 0, 4, 0, 0, 11],
              [0, 0, 17, 32, 11, 0]]

no_inicial = 0
no_final = 5

G = nx.from_numpy_array(np.array(distancias)) #Criando o grafo a partir de uma array
caminho_minimo = nx.dijkstra_path(G, no_inicial, no_final)
arcos_caminho = [(caminho_minimo[i], caminho_minimo[i+1]) for i in range(len(caminho_minimo)-1)] #Convertendo o caminho minimo para sequencia de laços
print(caminho_minimo, arcos_caminho)
```

```python
posicoes_dos_nos = {0:[0,1], 1:[1,2], 2:[2,0], 3:[3,2], 4:[3,0], 5:[5,1]} #Coordenadas dos nós

cor_dos_arcos = []
for arco in G.edges():
  if arco in arcos_caminho:
    cor_dos_arcos.append("red")
  else:
     cor_dos_arcos.append("green")

nx.draw_networkx(G, pos = posicoes_dos_nos, node_color = "white", edgecolors = "black", edge_color = cor_dos_arcos)

```
